### Visualize features

to get an intuition of the features

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bl = pd.read_csv('111409_w_63-BL1-081_bio.csv', sep = '\t') 
pa = pd.read_csv('111409_w_63-PA4-059_bio.csv', sep = '\t')
col_names = list(bl)[1:]
fig, ax = plt.subplots(figsize = (20,10), nrows = 2, ncols = 5, sharex=True, sharey='col')
for i in range(len(col_names)):     
    col = col_names[i]
    ax_bl = ax[0][i]
    ax_bl.plot(bl[col],'xkcd:sky blue')
    ax_bl.set_title('No pain - '+str(col))
    ax_pain = ax[1][i]
    ax_pain.plot(pa[col],'xkcd:orange')
    ax_pain.set_title('Highest Pain - '+str(col))

plt.show()

fig.savefig('vis_features.pdf')
```
